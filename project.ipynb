#### Exercise 10. Is the average of "Valdemorillo" and "Galapagar" price per square meter (price/m2) the same? (★★☆)

Print the both average prices and then write a conclusion about

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')


valdemorillo_df = ds[ds['level5'] == 'Valdemorillo'].copy()
galapagar_df = ds[ds['level5'] == 'Galapagar'].copy()


valdemorillo_df['pps'] = valdemorillo_df['price'] / valdemorillo_df['surface']
galapagar_df['pps'] = galapagar_df['price'] / galapagar_df['surface']


media_pps_valdemorillo = valdemorillo_df['pps'].mean()
media_pps_galapagar = galapagar_df['pps'].mean()


print(f"Media de price/m2 en Valdemorillo: {media_pps_valdemorillo}")
print(f"Media de price/m2 en Galapagar: {media_pps_galapagar}")

if media_pps_valdemorillo == media_pps_galapagar:
    print("Las medias de price/m2 son iguales.")
else:
    print("Las medias de price/m2 son diferentes.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv('assets/real_estate.csv', sep=';')

poblaciones_south_belt = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]


df_south_belt = ds[ds['level5'].isin(poblaciones_south_belt)]

median_prices = df_south_belt.groupby('level5')['price'].median()

plt.figure(figsize=(10, 6))
median_prices.plot(kind='bar', color='skyblue')
plt.title('Mediana de Precios en el "South Belt" de Madrid')
plt.xlabel('Población')
plt.ylabel('Mediana de Precio')
plt.grid(axis='y')
plt.show()
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

poblaciones_south_belt = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

df_south_belt = ds[ds['level5'].isin(poblaciones_south_belt)]

pivot_df = df_south_belt.pivot(columns='level5', values='price')

normalized_df = (pivot_df - pivot_df.min()) / (pivot_df.max() - pivot_df.min())

plt.figure(figsize=(10, 6))
normalized_df.plot.hist(bins=20, alpha=0.5, stacked=False)
plt.title('Histogramas de Precios por Población')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.legend(title='Población', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

poblaciones_south_belt = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

df_south_belt = ds[ds['level5'].isin(poblaciones_south_belt)]

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Scatter Plot de Precio por Metro Cuadrado para Poblaciones Seleccionadas')

for i, poblacion in enumerate(poblaciones_south_belt):
    df_poblacion = df_south_belt[df_south_belt['level5'] == poblacion]
    axs[i // 2, i % 2].scatter(df_poblacion['surface'], df_poblacion['price'], label=poblacion)
    axs[i // 2, i % 2].set_title(poblacion)
    axs[i // 2, i % 2].set_xlabel('Superficie')
    axs[i // 2, i % 2].set_ylabel('Precio')
    axs[i // 2, i % 2].legend()

# Ajusta el diseño y muestra el gráfico
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python

import pandas as pd
from ipyleaflet import Map, Marker, AwesomeIcon, basemaps


ds = pd.read_csv('assets/real_estate.csv', sep=';')

south_belt = ds[ds['level5'].isin(['Leganés', 'Fuenlabrada', 'Getafe', 'Alcorcón'])]


mapa = Map(center=(40.200772, -3.768001), zoom=10, min_zoom=1, max_zoom=20, basemap=basemaps.Esri.DeLorme)


icon1 = AwesomeIcon(name='home', marker_color='red', icon_color='darkred')
icon2 = AwesomeIcon(name='home', marker_color='green', icon_color='darkgreen')
icon3 = AwesomeIcon(name='home', marker_color='black', icon_color='gray')
icon4 = AwesomeIcon(name='home', marker_color='orange', icon_color='red')

icon_location = {
    'Leganés': icon1,
    'Fuenlabrada': icon2,
    'Getafe': icon3,
    'Alcorcón': icon4
}

for location in south_belt['level5'].unique():
    location_data = south_belt[south_belt['level5'] == location]

    for ind, row in location_data.iterrows():
        latitude = float(row['latitude'].replace(",", "."))
        longitude = float(row['longitude'].replace(",", "."))
        marker = Marker(location=(latitude, longitude), draggable=False, icon=icon_location[location])
        mapa.add_layer(marker)

mapa 
```
